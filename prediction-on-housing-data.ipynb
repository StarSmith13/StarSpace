```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
```

```python
iowa_filepath = '/kaggle/input/home-data-for-ml-course/train.csv' # this is your filepath, if you are working on JupyerNotebook just upload the file there, otherwise you can give the filepath on your computer
home_data = pd.read_csv(iowa_filepath)  
```

```python
y = home_data.SalePrice  # y is a Pandas Series, basically a unidimensional array holding values as per index
home_features = ['LotArea','GrLivArea','PoolArea']
X = home_data[home_features] # X is old school DataFrame
```

```python
home_model = DecisionTreeRegressor(random_state=1)
home_model.fit(X,y)
```
